```python
from pandas import DataFrame, Series, Index
age = Series(range(0,26+1))
index = Index(age)
years = Series(range(1987, 2012+2)) # :o)
uhh = Series([
 0,
 1,    # 1988
 1,
 3,    # 1990
 3,
 3,
 3,
 3,
 3,    # 1995
 3,
 3,
 3,
 3,
 3,    # 2000
 3,
 3,
 3,
 1,
 1,    # 2005
 1,
 1,
 1,
 1,
 1,    # 2010
 1,
 1,    # 2012,
 0
 ],
index=index)

df = DataFrame(index=index)
df['year'] = years
df['age'] = index
df['grade'] = df['age'] - 6

df['compulsory'] = ""
col = df['compulsory']
col[6] = "kinder"
col[11] = "elemen"
col[14] = "middle"
col[18] = "high"



df['college'] = ""
col = df['college']
col[20] = "assoc"
col[22] = "bach"
col[24] = "grad"
col[26] = "doct"

df['uhhhh'] = (df['compulsory'] + df['college']).shift(-1)
df['uhhhh'][3] = 'pres'
df['uhhhh'][4] = 'pres'
df['uhhhh'][23] = ''
df['uhhhh'][25] = ''
df['uhhhh'][26] = ''

col = df['age'] - 6
df['grade'] = col[col <= 12][col >= -2]
df['school'] = df['age'].apply(lambda x: x >= 3 and x!=24 and x<26 and '+' or '')
df['work'] = df['age'].apply(lambda x: x>=14 and x!=18 and x<24 and '+' or '')
df['abused'] = ""
df['abused'][14] = "attn skng friend; falsely"
df['abused'][16] = 'room with no window'
df['abused'][18] = 'early; undue'
df['abused'][21] = 'resigned'
df['abused'][22] = 'clean up after'
df['abused'][23] = 'kne, nck, shld, bck, eye'

df['uhh'] = uhh

df2 = df.copy()
df2['uhh'] = df['uhh'].apply(lambda x: '+'*x)

del df2['age']
del df2['uhhhh']
del df2['compulsory']
del df2['college']


display(uhh.plot(ylim=(0,5)))
display(df2)
```
