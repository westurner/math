```python
%load_ext sympyprinting
import sympy
from sympy import abc, Function, sin
from pandas import Series, Index, DataFrame
x = abc.x

f = ( x*sin(x**2)+1 )
print(type(f))
display(f)

def floatrange(start, stop, step=1):
    n = start
    while n < stop:
        yield n
        n += step

start=-10
end=10
index = Index(floatrange(start, end, 0.1))
series = Series( [ f.evalf(subs={x:n}) for n in index ], index=index)
#print(len(series))
series.plot()
```
