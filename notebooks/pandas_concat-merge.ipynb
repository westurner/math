```python
from pandas import DataFrame, Series, Index, concat, merge
```

```python
df1 = DataFrame(Series([10+n for n in range(10)],name='one'), index=Index(range(10)))
df2 = DataFrame(Series([20+n for n in range(10)],name='one'), index=Index(range(10)))
print(df1)
print(df2)

print('concat:outer') # default
dfconcat_outer = concat((df1, df2), join='outer')
print(dfconcat_outer)
print(len(dfconcat_outer))

print('concat:inner')
dfconcat_inner = concat((df1, df2), join='inner')
print(dfconcat_inner)
print(len(dfconcat_inner))

print('concat:ignore_index')
dfconcat_ignore_index = concat((df1, df2), ignore_index=True) 
print(dfconcat_ignore_index)
print(len(dfconcat_ignore_index))

print('merge:inner') # default
dfmerge_inner = merge(df1, df2, how='inner', )
print(dfmerge_inner)

print('merge:outer')
dfmerge_outer = merge(df1, df2, how='outer')
print(dfmerge_outer)

print('merge:left')
dfmerge_left = merge(df1, df2, how='left')
print(dfmerge_left)

print('merge:right')
dfmerge_right = merge(df1, df2, how='right')
print(dfmerge_right)
```

```python
fullseq = Series(range(0,10))
print(fullseq)
df1 = DataFrame(
        Series(fullseq+10,
            name='one'),
        index=Index(fullseq))
df2 = DataFrame(
        Series(fullseq+20,
            name='one'),
        index=Index(fullseq))
df3 = DataFrame(
        Series(fullseq[5:]+10,
            name='one'),
        index=Index(fullseq[5:]))
print(df1)
print(df2)
print(df3)
dfmerge = merge(df1, df2, how='outer')
dfmerge = merge(dfmerge, df3, how='outer')
print(dfmerge)
```
